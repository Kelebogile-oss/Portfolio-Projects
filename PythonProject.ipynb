#CLEANING UP THE **DATA**

```python
# only do USA Races, 50km or 50mi, 2020

#STEP 1  SHOW 50 MILES OR 50KM
#fix the spaces inbetween the columns first before using the 50MILES or 50KM
df[df['Event distance/length']=='50km']
```

```python
# Combine 50Km /50mi  with "ISIN" DataFrame || 'isisn'= method in pandas used to  filter DataFrame by checking whether each element in then DataFrame is contained in a Specificied list or iterable.
df[df['Event distance/length'].isin(['50km','50mi'])]
```

```python
# Combining the Event distance  and the Event year.
df[(df['Event distance/length'].isin(['50km','50mi']))&(df['Year of event'] == 2020)]
```

```python
#REMANE COLUMNS

#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#Athlete_Age                    int64
#dtype: object
```

```python
data2 = data2.rename(columns={ 'Year of event':'year',
                               'Event dates':'race_day',
                               'Event name':'race_name',
                               'Event distance/length':'race_distance',
                               'Event number of finishers':'race_number_of_finishers',
                               'Athlete performance':'athlete_performance',
                               'Athlete gender':'athlete_gender',
                               'Athlete average speed':'athlete_average_speed',
                               'Athlete ID':'athlete_id',
})
```

```python
data2.head()
```

```python
#reorder columns

data3 = data2[['race_day','race_name','race_distance','race_number_of_finishers','athlete_id','athlete_gender','Athlete_Age','athlete_performance','athlete_average_speed','year']]
```

```python
data3.head()
```

# **Analysing and Visualising Data**

```python
sns.histplot(data3['race_distance'])
plt.title('The Number of Race in 50 km and 50mi category')
plt.show()
```

```python
##The Number of Race in 50 km and 80 km category by gender
sns.histplot(data3,  x= 'race_distance', hue = 'athlete_gender',palette = 'tab10')
plt.title('The Number of Race in 50 km and 80 km category By gender')
plt.show()
```

```python
#The relationship between the race distance and the race speed
sns.distplot(data3[data3['race_distance'] == '50mi']['athlete_average_speed'])
plt.title('Relationship between the race_distance and the athlete_average_speed on 50mi')
plt.show()
```

```python
#Relationship between the distance of the race and the speed of the athlete

plt.figure(figsize=(10,6))
sns.violinplot(data3, x='race_distance', y='athlete_average_speed', hue='athlete_gender', split=True, inner='quartile')
plt.tight_layout()
plt.title('Relationship between the distance of the race and the athlete average speed')
plt.xlabel('Race_distance')
plt.ylabel('Athlete_Average_Speed')
plt.legend(title='Athlete Gender')
plt.show()
```

```python
# the relationship between athlete age and athelete average speed
sns.lmplot(data = data3, x='Athlete_Age', y='athlete_average_speed', hue = 'athlete_gender',palette = 'tab10')
plt.title('The relationship between athlete age and athelete average speed',fontsize = 12)
plt.show()
```

```python
#Find the differen  speed between the genders

data3.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
#  What Age groups are the best in the 50mi  Race (20 + race min)(show 15)
data3.query('race_distance == "50mi"').groupby('Athlete_Age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
#  What Age groups are the worst in the 50mi  Race (20 + race min)(show 15)
data3.query('race_distance == "50mi"').groupby('Athlete_Age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19')
```

```python
#Convert date type into the correct format. Currently its DD.MM.YYYY.
data3['race_month'] = data3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
# applying a lambda function to get the season of the event
data3['race_season'] = data3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
data3.head()
```

```python
#comparing the average speed by season
data3 = data3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values(by='mean', ascending=False)
data3
```

```python
# @title Mean X and Count over Time

sns.lineplot(x = 'race_season', y = 'mean', data = data3)
sns.lineplot(x = 'race_season', y = 'count', data = data3)
```
